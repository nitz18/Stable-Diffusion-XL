```python
from huggingface_hub.repocard import RepoCard
from diffusers import DiffusionPipeline
import torch

# lora_model_id = <"lora-sdxl-dreambooth-id">
# card = RepoCard.load(lora_model_id)
# base_model_id = card.data.to_dict()["base_model"]

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16)
pipe = pipe.to("cuda")
pipe.load_lora_weights("/content/diffusers/examples/dreambooth/nithyashreemanohar/Dreambooth_LoRa/checkpoint-200/pytorch_lora_weights.safetensors")
```

```python
images = pipe("A picture of Ronaldo showing the thumbs up gesture on the moon", num_inference_steps=50).images[0]
images.save("Ronaldo_200_moon.png")
```

```python
image = pipe("A picture of Ronaldo showing the thumbs up gesture on top of a hill", num_inference_steps=50).images[0]
image.save("Ronaldo_200__hill.png")
```
